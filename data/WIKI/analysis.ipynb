```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 100)

# Read data
data_folder = './'
train = pd.read_csv(data_folder + 'train', header=None, sep='\t')
test = pd.read_csv(data_folder + 'test', header=None, sep='\t')
valid = pd.read_csv(data_folder + 'valid', header=None, sep='\t')
```

```python
len(train), len(test), len(valid)
```

```python
def split_rel_att(df):
    rel_query = df.loc[df[2].str.startswith('Q')].reset_index(drop=True)
    att_query = df.loc[~df[2].str.startswith('Q')].reset_index(drop=True)
    rel_query.columns = ['subject', 'relation', 'object', 'timestamp']
    att_query.columns = ['subject', 'attribute', 'value', 'timestamp']
    att_query['value'] = att_query['value'].astype(float)
    return rel_query, att_query
```

```python
train_rel, train_att = split_rel_att(train)
test_rel, test_att = split_rel_att(test)
valid_rel, valid_att = split_rel_att(valid)
```

```python
def count_attribute_occurrences(df, col='attribute'):
    return df.groupby(col).size().reset_index(name='count')

train_att_count = count_attribute_occurrences(train_att, 'attribute')
test_att_count = count_attribute_occurrences(test_att, 'attribute')
valid_att_count = count_attribute_occurrences(valid_att, 'attribute')

att_count = train_att_count.merge(test_att_count, on='attribute', how='outer', suffixes=('_train', '_test'))
att_count = att_count.merge(valid_att_count, on='attribute', how='outer')
att_count.columns = ['attribute', 'train', 'test', 'valid']
att_count = att_count.fillna(0)
att_count['train'] = att_count['train'].astype(int)
att_count['test'] = att_count['test'].astype(int)
att_count['valid'] = att_count['valid'].astype(int)
att_count = att_count.sort_values('test', ascending=False).reset_index(drop=True)
att_count
```

```python
train_rel_count = count_attribute_occurrences(train_rel, 'relation')
test_rel_count = count_attribute_occurrences(test_rel, 'relation')
valid_rel_count = count_attribute_occurrences(valid_rel, 'relation')

rel_count = train_rel_count.merge(test_rel_count, on='relation', how='outer', suffixes=('_train', '_test'))
rel_count = rel_count.merge(valid_rel_count, on='relation', how='outer')
rel_count.columns = ['relation', 'train', 'test', 'valid']
rel_count = rel_count.fillna(0)
rel_count['train'] = rel_count['train'].astype(int)
rel_count['test'] = rel_count['test'].astype(int)
rel_count['valid'] = rel_count['valid'].astype(int)
rel_count = rel_count.sort_values('test', ascending=False).reset_index(drop=True)
rel_count
```
